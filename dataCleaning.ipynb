```python
import pandas as pd
#First of, we want to create a new attribute that shows the week number, based on the date as we will use this later

# Read the Excel file into a DataFrame
df = pd.read_excel('NyDataSet.xlsx')

# Convert the 'week' column to datetime format
df['week'] = pd.to_datetime(df['week'])

# Add a new column 'week_number' with the ISO week number
df['week_number'] = df['week'].dt.isocalendar().week

# Save the updated DataFrame to a new Excel file
df.to_excel('SpofityData.xlsx', index=False)

```

```python
#Here we 
import pandas as pd

df = pd.read_excel('SpotifyData.xlsx')

# Remove duplicates based on 'id' but keep entries from different weeks
filtered_df = df.drop_duplicates(subset=['uri', 'week_number'])

# Save the filtered DataFrame to a new Excel file
filtered_df.to_excel('output_file_filtered.xlsx', index=False)
```

```python
filtered_df
```

```python
#Check to see whether there are any rows where the combination of a specific uri and a weeknumber is present multiple times
duplicate_rows = filtered_df[filtered_df.duplicated(subset=['uri', 'week_number'], keep=False)]

# Print rows with duplicates
if not duplicate_rows.empty:
    print("Rows with duplicate 'uri' and 'week_number':")
    print(duplicate_rows)
else:
    print("No duplicates found.")
```
