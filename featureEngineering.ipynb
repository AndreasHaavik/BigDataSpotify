```python
# create dummy values for genre
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Read the Excel file into a DataFrame
clean_df=pd.read_excel('SpotifyDataClean.xlsx')

# Flatten the list of genres into a single series
all_genres = clean_df['artist_genre'].str.split(', ').apply(pd.Series).stack().reset_index(drop=True)

# Initialize the Label Encoder
label_encoder = LabelEncoder()

# Fit the label encoder and transform genres to numeric codes
encoded_genres = label_encoder.fit_transform(all_genres)

# Mapping the original genres to their numeric codes
genre_mapping = dict(zip(all_genres, encoded_genres))

# Replace the genres in the original dataframe with their corresponding codes
clean_df['encoded_genres'] = clean_df['artist_genre'].apply(lambda x: [genre_mapping[genre] for genre in x.split(', ')])
```

```python
clean_df['encoded_genres']
```

```python
# Removing duplicate values within each list in the 'encoded_genres' field
clean_df['encoded_genres'] = clean_df['encoded_genres'].apply(lambda x: list(set(x)))

# Displaying the first few rows to verify the changes
clean_df[['artist_genre', 'encoded_genres']].head()
```

```python
# Extracting unique genres from the 'artist_genre' column
unique_genres = set(genre for genre_list in clean_df['artist_genre'].dropna() for genre in genre_list.split(', '))

# Initialize the Label Encoder
label_encoder = LabelEncoder()

# Fit the label encoder and transform genres to numeric codes
label_encoder.fit(list(unique_genres))
encoded_genres_unique = label_encoder.transform(list(unique_genres))

# Creating a mapping dictionary
genre_encoding_map = dict(zip(list(unique_genres), encoded_genres_unique))

# Displaying the genre to encoded number mapping
genre_encoding_map
```
